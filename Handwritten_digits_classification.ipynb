```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
(X_train, y_train),(X_test, y_test)=keras.datasets.mnist.load_data()
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_test_flattened=X_test.reshape(len(X_test),28*28)

X_train_flattened.shape
```

```python
X_test_flattened.shape
```

```python
model=keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train_flattened,y_train, epochs=10)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
model.predict(X_test_flattened)
```

```python
y_pred=model.predict(X_test_flattened)
```

```python
np.argmax(y_pred[0])
```

```python
plt.matshow(X_test[0])
```

```python
y_test[:5]
```

```python
y_pred_labels=[np.argmax(i) for i in y_pred]
y_pred_labels[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test, predictions=y_pred_labels)
```

```python
import seaborn as sn
plt.figure(figsize=(10,8))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
```

```python
model2=keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model2.fit(X_train_flattened, y_train, epochs=5)
```

```python
y_pred_2=model2.predict(X_test_flattened)
```

```python
y_pred_label2=[np.argmax(i) for i in y_pred_2]
y_pred_label2[:5]
```

```python
cm2=tf.math.confusion_matrix(labels=y_test, predictions=y_pred_label2)
import seaborn as sn
plt.figure(figsize=(10,8))
sn.heatmap(cm2, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
```

```python
model3=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')

])
model3.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model3.fit(X_train, y_train, epochs=5)
```

```python
y_pred_3=model3.predict(X_test)
```

```python
y_pred_label3=[np.argmax(i) for i in y_pred_3]
y_pred_label3[:5]
```

```python
cm3=tf.math.confusion_matrix(labels=y_test, predictions=y_pred_label3)
import seaborn as sn
plt.figure(figsize=(10,8))
sn.heatmap(cm3, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
```
